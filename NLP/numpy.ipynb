# 数组创建

```python
import numpy as np

# 从Python列表创建
arr = np.array([1,2,3,4,5]) 

# 创建全0或全1数组 (常用于初始化权重或偏置)
zeros_arr = np.zeros((2,3)) # 2行3列的全0数组
ones_arr = np.ones((3,2)) # 3行2列的全1数组

# 创建指定值的数组
full_arr = np.full((2,2),7) # 2行2列的数组，所有元素都是7

# 创建单位矩阵 (在特定初始化中会用到)
eye_arr = np.eye(3) # 3x3的单位矩阵

# 创建随机数组 (权重初始化的核心)
rand_arr = np.random.rand(2,2) # 2行2列的随机数数组，元素在[0,1)之间
randint_arr = np.random.randint(1,10,size=(2,2)) # 2行2列的随机整数数组，元素在[1,10)之间
```

# 数组属性

```python
arr = np.array([[1,2,3],[4,5,6]]) # 2行3列的二维数组
print(arr.shape) # 输出: (2, 3),表示数组的维度
print(arr.ndim) # 输出: 2,表示数组的维数
print(arr.size) # 输出: 6,表示数组中元素的总数
print(arr.dtype) # 输出: int64,表示数组元素的数据类型
```

# 索引与切片

```python
arr = np.arrange(12).reshape(3,4) # 创建一个包含0-11的数组，并重塑为3行4列

# 基本索引
print(arr[0,1]) # 输出: 1,访问第一行第二列的元素
print(arr[1,:]) # 输出: [4 5 6 7],访问第二行的所有元素
print(arr[:,2]) # 输出: [ 2  6 10],访问所有行的第三列元素

# 切片
print(arr[:2,:3]) # 输出: [[0 1 2] [4 5 6]],访问前两行和前三列的子数组

# 布尔索引 
print(arr[arr>5]) # 输出: [ 6  7  8  9 10 11],访问数组中大于5的元素
```

# 数组运算

```python
arr = np.array([1,2,3])

# 元素级运算
print(arr*2) # 输出: [2 4 6],每个元素乘以2
print(arr+5) # 输出: [6 7 8],每个元素加上5
print(arr**2) # 输出: [1 4 9],每个元素的平方
print(np.sqrt(arr)) # 输出: [1. 1.41421356 1.73205081],每个元素的平方根
```

# 广播机制

```python
matrix = np.array([[1,2,3],[4,5,6]])
vector = np.array([10,20,30])

# 向量会被“广播”成 (2, 3) 的矩阵，然后进行加法
result = matrix + vector # 输出: [[11 22 33] [14 25 36]],矩阵的每一行都加上了向量
```

# 矩阵乘法

```python
A = np.array([[1,2],[3,4]]) # 2x2矩阵
B = np.array([[5,6],[7,8]]) # 2x2矩阵

# 方式1: 使用 @ 运算符 (推荐)
C = A @ B 

# 方式2: 使用 np.matmul 函数
C = np.matmul(A,B)

# 输出: [[19 22] [43 50]],矩阵A和B的乘积
```

# 重塑与转置

```python
arr = np.arange(6) # [0 1 2 3 4 5]

# 重塑(reshape)
reshaped = arr.reshape(2,3) # [[0 1 2] [3 4 5]],将一维数组重塑为2行3列的二维数组

# 转置(transpose)
transposed = reshaped.T # [[0 3] [1 4] [2 5]],将数组的行和列进行交换
```

# 常用统计函数

```python
scores = np.array([82,90,78,88,75])

print(np.mean(scores)) # 输出: 82.6,计算平均值
print(np.sum(scores)) # 输出: 413,计算总和
print(np.std(scores)) # 输出: 5.1478150704935,计算标准差
print(np.max(scores)) # 输出: 90,计算最大值
print(np.min(scores)) # 输出: 75,计算最小值
```

# 数组拼接与分割

```python
arr1 = np.array([[1,2],[3,4]])
arr2 = np.array([[5,6],[7,8]])

# 纵向拼接 (增加行)
v_stack = np.vstack((arr1,arr2)) # [[1 2] [3 4] [5 6] [7 8]]

# 横向拼接 (增加列)
h_stack = np.hstack((arr1,arr2)) # [[1 2 5 6] [3 4 7 8]]
```
