# 2. Decoder-only架构
*   **特点**：这是当前主流大语言模型（如 GPT、LLaMA、Qwen）的绝对主导架构。它采用自回归方式，即根据上文逐个预测下一个词元（token），天生适合文本生成。
*   **优势**：训练目标简单（预测下一个词），易于扩展到超大规模。推理时可复用 KV Cache，在多轮对话等场景下效率极高。
*   **代表模型**：GPT-4, LLaMA 3, Qwen 2.5, Mistral。
![Decoder-only架构](../image/Decoder-only架构.png)

# 3. Encoder-Only架构
*   **特点**：能够“看到”完整的输入文本，进行双向上下文理解，但不擅长生成式任务。
*   **优势**：在文本分类、情感分析、信息提取等需要深度理解的任务上表现卓越。
*   **代表模型**：BERT, RoBERTa。
![Encoder-Only架构](../image/Encoder-Only架构.png)

# 4. Encoder-Decoder架构
*   **特点**：结合了前两者的优点。编码器负责理解输入，解码器负责生成输出。
*   **优势**：非常适合“序列到序列”的转换任务，如机器翻译、文本摘要等。
*   **代表模型**：T5, BART。
