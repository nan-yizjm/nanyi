#  基础设置：解决中文与显示问题

```python
import matplotlib.pyplot as plt
import numpy as np

# --- 核心配置：解决中文显示问题 ---
plt.rcParams["font.sans-serif"] = ["SimHei"] # 设置中文字体为黑体
plt.rcParams["axes.unicode_minus"] = False # 解决负号显示问题

# --- 核心配置：Jupyter Notebook 显示设置 ---
# 如果在 Jupyter 中，加上这行代码可以让图表内嵌显示
# %matplotlib inline 
```

# 训练曲线可视化

```python
# 模拟训练数据
steps = np.arange(0,1000,10) 
loss = 1/(steps+1)+np.random.rand(len(steps))*0.05 # 模拟损失曲线，带有一些随机噪声

# 1. 创建画布 (figsize: 宽, 高)
plt.figure(figsize=(10,6)) # 创建一个 10x6 英寸的画布

# 2. 绘制折线图
plt.plot(steps,loss,label='Training Loss',color='blue',alpha=0.6) # 绘制折线图，设置标签、颜色和透明度

# 3. 平滑处理 (移动平均) - LLM 训练中常用技巧
window_size = 50    # 移动平均窗口大小
smooth_loss = np.convolve(loss,np.ones(window_size)/window_size,mode='valid') # 计算移动平均
smooth_steps = steps[window_size-1:] # 平滑后的步数对应关系

plt.plot(smooth_steps,smooth_loss,label='Smoothed Loss',color='red',linewidth=2) # 绘制平滑后的曲线，设置标签、颜色和线宽

# 4. 添加标签和标题
plt.title('LLM Training Loss Curve',fontsize=16) # 设置标题和字体大小
plt.xlabel('Training Steps',fontsize=12) # 设置 x 轴标签和字体大小
plt.ylabel('Loss Value',fontsize=12) # 设置 y 轴标签和字体大小

# 5. 显示图例和网格
plt.legend(loc='best') # 显示图例，自动选择最佳位置
plt.grid(True,linestyle='--',alpha=0.5) # 显示网格，设置线型和透明度

# 6. 保存与显示
plt.savefig('training_loss.png',dpi=300,bbox_inches='tight') # 保存图像，设置分辨率和边距
plt.show() # 显示图像
```

# 数据分布与对比

```python
fig,axes = plt.subplots(1,2,figsize=(15,5)) # 创建一个包含 1 行 2 列子图的画布,设置整体大小

# --- 左图：文本长度分布 (直方图) ---
# 模拟 1000 个样本的文本长度
text_lengths = np.random.normal(loc=512,scale=100,size=1000) # 创建一个正态分布的文本长度数据，平均值为 512，标准差为 100

axes[0].hist(text_lengths,bins=30,color='skyblue',edgecolor='black',alpha=0.7) # 绘制直方图，设置颜色、边框颜色和透明度
axes[0].set_title('Distribution of Text Lengths') # 设置标题
axes[0].set_xlabel('Length(Tokens)') # 设置 x 轴标签
axes[0].set_ylabel('Frequency') # 设置 y 轴标签
axes[0].axvline(512,color='red',linestyle='dashed',linewidth=1,label='Max Length Limit') # 添加垂直线表示最大长度限制，设置颜色、线型、线宽和标签
axes[0].legend() # 显示图例

# --- 右图：类别样本数量 (柱状图) ---
categories = ['Positive','Negative','Neutral'] # 定义类别
counts = [500,200,300] # 模拟每个类别的样本数量

axes[1].bar(categories,counts,color=['green','red','gray']) # 绘制柱状图，设置颜色
axes[1].set_title('Class Distribution') # 设置标题
axes[1].set_xlabel('Sentiment') # 设置 x 轴标签
axes[1].set_ylabel('Count') # 设置 y 轴标签

# 在柱子上添加数值标签
for i,v in enumerate(counts): # 遍历每个柱子和对应的数值
   axes[1].text(i,v+5,str(v),ha='center') # 在柱子上方添加数值标签，设置水平对齐方式为居中,v+5 是为了让标签稍微高于柱子

plt.tight_layout() # 调整子图间距
plt.show() # 显示图像
```

# 注意力热力图

```python
# 模拟一个 10x10 的注意力矩阵 (Attention Matrix)
# 假设是 "The cat sat on the mat" 这句话的自注意力权重
words = ['The','cat','sat','on','the','mat','[PAD]','[PAD]','[PAD]','[PAD]'] # 模拟一个包含 10 个词的句子，后面用 [PAD] 填充
attention_weights = np.random.rand(10,10) # 创建一个 10x10 的随机注意力权重矩阵
# 模拟一些高权重区域
attention_weights[1,5] = 0.9 # "cat" 对 "mat" 的注意力权重较高
attention_weights[0,1] = 0.8 # "The" 对 "cat" 的注意力权重较高

plt.figure(figsize=(8,6)) # 创建一个 8x6 英寸的画布
# imshow 是画热力图的核心函数
plt.imshow(attention_weights,cmap='viridis',interpolation='nearest') # 绘制热力图，设置颜色映射和插值方式

# 添加颜色条
plt.colorbar(label='Attention Weight')  

# 设置坐标轴标签
plt.xticks(np.arange(10),words,rotation=45) # 设置 x 轴刻度和标签，旋转标签以避免重叠
plt.yticks(np.arange(10),words) # 设置 y 轴刻度和标签

# 在格子里写上数值 (可选)
for i in range(10):
    for j in range(10):
        plt.text(j,i,f'{attention_weights[i,j]:.2f}',ha='center',
                 va='center',color='white' if attention_weights[i,j]>0.5 else 'black') # 在每个格子中添加数值标签，设置水平和垂直对齐方式，根据权重值选择文本颜色以确保可读性
        
plt.title('Self-Attention Map Visualization') # 设置标题
plt.show()
```

# 面向对象绘图与子图

```python
# 创建 2x2 的网格布局
fig,axs = plt.subplots(2,2,figsize=(15,10)) # 创建一个包含 2 行 2 列子图的画布，设置整体大小

# 子图 1: 训练集 Loss
axs[0,0].plot([1,2,3],[1,2,3],label='Train') # 在第一个子图中绘制训练集的损失曲线，设置标签为 'Train'
axs[0,0].set_title('Train Loss') # 设置第一个子图的标题
axs[0,0].legend() # 显示图例

# 子图 2: 验证集 Loss
axs[0,1].plot([1,2,3],[1.5,2.5,3.5],label='Val',color='orange') # 在第二个子图中绘制验证集的损失曲线，设置标签为 'Val' 和颜色
axs[0,1].set_title('Validation Loss') # 设置第二个子图的标题
axs[0,1].legend() # 显示图例

# 子图 3: 学习率变化
axs[1,0].plot([1,2,3],[0.001,0.0005,0.0001],label='LR') # 在第三个子图中绘制学习率的变化曲线，设置标签为 'LR'
axs[1,0].set_title('Learing Rate Schedule') # 设置第三个子图的标题
axs[1,0].legend() # 显示图例

# 子图 4: 评估指标
axs[1,1].bar(['BLEU','ROUGE'],[0.8,0.75]) # 在第四个子图中绘制评估指标的柱状图，设置 x 轴标签和对应的数值
axs[1,1].set_title('Evaluation Metrics') # 设置第四个子图的标题
axs[1,1].set_ylim(0,1) # 设置 y 轴范围为 0 到 1

plt.suptitle('Experiment Report: Model v1.0',fontsize=20) # 设置整个图表的标题和字体大小
plt.tight_layout() # 调整子图间距，避免标题和子图重叠
plt.show() # 显示图像
```
