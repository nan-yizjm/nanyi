#  基础设置：解决中文与显示问题

```python
import matplotlib.pyplot as plt
import numpy as np

# --- 核心配置：解决中文显示问题 ---
plt.rcParams["font.sans-serif"] = ["SimHei"] # 设置中文字体为黑体
plt.rcParams["axes.unicode_minus"] = False # 解决负号显示问题

# --- 核心配置：Jupyter Notebook 显示设置 ---
# 如果在 Jupyter 中，加上这行代码可以让图表内嵌显示
# %matplotlib inline 
```

# 训练曲线可视化

```python
# 模拟训练数据
steps = np.arange(0,1000,10) 
loss = 1/(steps+1)+np.random.rand(len(steps))*0.05 # 模拟损失曲线，带有一些随机噪声

# 1. 创建画布 (figsize: 宽, 高)
plt.figure(figsize=(10,6)) # 创建一个 10x6 英寸的画布

# 2. 绘制折线图
plt.plot(steps,loss,label='Training Loss',color='blue',alpha=0.6) # 绘制折线图，设置标签、颜色和透明度

# 3. 平滑处理 (移动平均) - LLM 训练中常用技巧
window_size = 50    # 移动平均窗口大小
smooth_loss = np.convolve(loss,np.ones(window_size)/window_size,mode='valid') # 计算移动平均
smooth_steps = steps[window_size-1:] # 平滑后的步数对应关系

plt.plot(smooth_steps,smooth_loss,label='Smoothed Loss',color='red',linewidth=2) # 绘制平滑后的曲线，设置标签、颜色和线宽

# 4. 添加标签和标题
plt.title('LLM Training Loss Curve',fontsize=16) # 设置标题和字体大小
plt.xlabel('Training Steps',fontsize=12) # 设置 x 轴标签和字体大小
plt.ylabel('Loss Value',fontsize=12) # 设置 y 轴标签和字体大小

# 5. 显示图例和网格
plt.legend(loc='best') # 显示图例，自动选择最佳位置
plt.grid(True,linestyle='--',alpha=0.5) # 显示网格，设置线型和透明度

# 6. 保存与显示
plt.savefig('training_loss.png',dpi=300,bbox_inches='tight') # 保存图像，设置分辨率和边距
plt.show() # 显示图像
```

# 数据分布与对比

```python
fig,axes = plt.subplots(1,2,figsize=(15,5)) # 创建一个包含 1 行 2 列子图的画布,设置整体大小

# --- 左图：文本长度分布 (直方图) ---
# 模拟 1000 个样本的文本长度
text_lengths = np.random.normal(loc=512,scale=100,size=1000) # 创建一个正态分布的文本长度数据，平均值为 512，标准差为 100

axes[0].hist(text_lengths,bins=30,color='skyblue',edgecolor='black',alpha=0.7) # 绘制直方图，设置颜色、边框颜色和透明度
axes[0].set_title('Distribution of Text Lengths') # 设置标题
axes[0].set_xlabel('Length(Tokens)') # 设置 x 轴标签
axes[0].set_ylabel('Frequency') # 设置 y 轴标签
axes[0].axvline(512,color='red',linestyle='dashed',linewidth=1,label='Max Length Limit') # 添加垂直线表示最大长度限制，设置颜色、线型、线宽和标签
axes[0].legend() # 显示图例
```
